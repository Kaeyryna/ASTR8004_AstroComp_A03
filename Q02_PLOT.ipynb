```python
#KAEYRYNA ARYSSA BINTI KAEYDA ASTR8004 ASSIGNMENT 03

import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np

plt.rcParams['font.size'] = 12

colmag = pd.read_csv('2.6_plot1_CMD-result.csv')
tmass = pd.read_csv('2.6_plot2_J-KvsH-result.csv')

#data to plot CMD
bprp = np.array(colmag['bp_rp'])
abs_MG = colmag['phot_g_mean_mag'] + 5 - 5*np.log10(1000/ colmag['parallax'])

#data to plot from tmass
j_k = tmass['j_m'] - tmass['ks_m']
arr_j_k = np.array(j_k)
hmag = np.array(tmass['h_m'])

#create the figure to plot 2 panels
fig, axs = plt.subplots(1,2, figsize=(10, 5))

#Panel 1
ax = axs[0]
ax.plot(bprp, abs_MG, 'o', markersize=2, alpha=0.5, color='#0C7BDC')
ax.set_title('Color-Magnitude Diagram (CMD) of\n Gaia BP-RP vs absolute G magnitude')
ax.set_ylabel('Absolute G magnitude (mag)')
ax.set_xlabel('BP-RP (mag)')
ax.invert_yaxis()

#Panel 2
ax = axs[1]
ax.plot(arr_j_k, hmag, 'o', markersize=2, alpha=0.5, color='#FFC20A')
ax.set_title('2MASS J-K vs apparent H magnitude')
ax.set_ylabel('Apparent H magnitude (mag)')
ax.set_xlabel('J-K (mag)')
ax.invert_yaxis()

# Show plot
plt.tight_layout()
plt.savefig('cmds_R147.png', dpi=200, bbox_inches='tight')
plt.show()
```
